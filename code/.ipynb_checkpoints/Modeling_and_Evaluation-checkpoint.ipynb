```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_log_error, mean_squared_error, mean_absolute_error
```

```python
train_data = pd.read_csv('../datasets/train_data_clean.csv')
train_data.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
train_data.head()
```

```python
test_data = pd.read_csv('../datasets/test_data_clean.csv')
test_data.drop(columns = 'Unnamed: 0', inplace = True)
```

### Baseline model

```python
train_data['baseline'] = train_data['SalePrice'].mean()
```

```python
resb = train_data['baseline']
y_act = np.array(train_data['SalePrice'])
y_hat = np.array(train_data['baseline'])
print(f'Sum Squared error {sum(resb**2)}')
print(f'Mean absolute error {mean_absolute_error(y_act, y_hat)}')
print(f'Mean Squared Error {mean_squared_error(y_act, y_hat)}')
print(f'Root Mean Squared Error {np.sqrt(mean_squared_error(y_act, y_hat))}')
```

```python
X = train_data_model.drop(columns=['Id','SalePrice','total_sqr_ft^2','Garage Area'])
y = train_data_model['SalePrice']
```

```python
testsub = test_data_model.drop(columns='Id')
```

### Split train and test 

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state = 42)
```

# **Preprocessing and Modeling**

I am going to use a OLS (orninaty least squares), as know as linear regression, to find the best fitting line to predict the Saling price of a home. The model works by minimizing sum of the squared residuals. 

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```
